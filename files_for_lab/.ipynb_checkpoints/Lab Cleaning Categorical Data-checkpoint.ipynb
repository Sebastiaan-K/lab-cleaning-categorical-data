```python
def split_dataframe(dataframe):
    numerical_columns = dataframe.select_dtypes(include=np.number)
    categorical_columns = dataframe.select_dtypes(exclude=np.number)
    return numerical_columns, categorical_columns

numerical_df, categorical_df = split_dataframe(customer_df)
```

```python
def split_numerical_continuous_discrete(dataframe, threshold=36):
    continuous_columns = dataframe.loc[:, dataframe.nunique() > threshold]
    discrete_columns = dataframe.loc[:, dataframe.nunique() <= threshold]
    return continuous_columns, discrete_columns

continuous_df, discrete_df = split_numerical_continuous_discrete(numerical_df)

```

```python
def plot_discrete(dataframe):
    fig, axes = plt.subplots(nrows=len(dataframe.columns), figsize=(16, 16))
    for ax, column in zip(axes, dataframe.columns):
        sns.barplot(x=dataframe[column].value_counts().index, y=dataframe[column].value_counts(), ax=ax)
        ax.set_title(column)
    plt.tight_layout()
    plt.show()

plot_discrete(discrete_df)
```

```python
def plot_continuous(dataframe):
    fig, axes = plt.subplots(nrows=len(dataframe.columns), figsize=(16, 16))
    for ax, column in zip(axes, dataframe.columns):
        sns.histplot(dataframe[column], ax=ax, kde=True)
        ax.set_title(column)
    plt.tight_layout()
    plt.show()

plot_continuous(continuous_df)
```

```python
def plot_outliers(dataframe):
    fig, axes = plt.subplots(nrows=len(dataframe.columns), figsize=(16, 16))
    for ax, column in zip(axes, dataframe.columns):
        sns.boxplot(x=dataframe[column], ax=ax)
        ax.set_title(column)
    plt.tight_layout()
    plt.show()

plot_outliers(continuous_df)
```

1. Define a function that creates seaborn countplots for each categorical column in a pandas DataFrame, sorting the bars by frequency and placing the bars on the y-axis if there are six or more unique values.

```python
def plot_categorical_count(df):
    categorical_cols = df.select_dtypes(include=['object']).columns

    for col in categorical_cols:
        fig, ax = plt.subplots(figsize=(10, 6))
        order = df[col].value_counts().index
        if df[col].nunique() < 6:
            sns.countplot(x=col, data=df, order=order, ax=ax)
        else:
            sns.countplot(y=col, data=df, order=order, ax=ax)
        plt.show()
```

13. Define a function to remove outliers from a numerical continuous column based on a threshold of standard deviations.

```python
def remove_outliers(df, column, thr=3):
    mean = df[column].mean()
    std = df[column].std()
    upper_limit = mean + thr * std
    lower_limit = mean - thr * std
    return df[(df[column] <= upper_limit) & (df[column] >= lower_limit)]
```
